```python
import xgboost as xgb
from sklearn.metrics import root_mean_squared_error
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import joblib
```

```python
X_train = pd.read_pickle("../dataset/processed_data/rf/Xrf_train.pkl")
X_test = pd.read_pickle("../dataset/processed_data/rf/Xrf_test.pkl")
y_train = pd.read_pickle("../dataset/processed_data/rf/yrf_train.pkl")
y_test = pd.read_pickle("../dataset/processed_data/rf/yrf_test.pkl")
```

```python
X_train = np.array(X_train)
X_test = np.array(X_test)
y_train = np.array(y_train)
y_test = np.array(y_test)
```

```python
xgb_model = xgb.XGBRegressor(
    objective='reg:squarederror',  # Standard loss function for regression
    n_estimators=1000,  # Number of boosting rounds
    learning_rate=0.05,  # Step size shrinkage to prevent overfitting
    max_depth=10,  # Maximum depth of trees
    subsample=0.8,  # Fraction of samples used for training
    colsample_bytree=0.8,  # Fraction of features used per tree
    random_state=42
)
```

```python
xgb_model.fit(X_train, y_train)
```

```python
y_pred = xgb_model.predict(X_test)
```

```python
rmse = root_mean_squared_error(y_test, y_pred)
print(f"âœ… XGBoost Model RMSE: {rmse:.2f}")
```

```python
margen_error = (rmse / y_test.mean()) * 100
print(margen_error)
```

```python
from sklearn.model_selection import GridSearchCV

param_grid = {
    'n_estimators': [100, 500, 1000], 
    'learning_rate': [0.01, 0.05, 0.1],  
    'max_depth': [4, 6, 8]
}

grid_search = GridSearchCV(
    xgb.XGBRegressor(objective='reg:squarederror', random_state=42), 
    param_grid, 
    cv=5, 
    scoring='neg_root_mean_squared_error', 
    n_jobs=-1
)

grid_search.fit(X_train, y_train)

# Best parameters

# Train with best params
best_xgb = xgb.XGBRegressor(**grid_search.best_params_, objective='reg:squarederror', random_state=42)
best_xgb.fit(X_train, y_train)

# Predict and evaluate
y_pred = best_xgb.predict(X_test)
rmse = root_mean_squared_error(y_test, y_pred)

```

```python
print(f"âœ… Best Parameters: {grid_search.best_params_}")
```

```python
print(f"âœ… Optimized XGBoost Model RMSE: {rmse:.2f}")
```

```python
margen_error = (rmse / y_test.mean()) * 100
print(margen_error)
```

```python


# Plot feature importance
xgb.plot_importance(best_xgb, max_num_features=20, importance_type='weight', height=0.5)
plt.title("Feature Importance in XGBoost")
plt.show()
```

```python
#--------------DecisionTree-------------
```

```python
from sklearn import tree
```

```python
decisionTree_model = tree.DecisionTreeRegressor()
decisionTree_model.fit(X_train, y_train)
```

```python
predicciones_dtr = decisionTree_model.predict(X_test)
```

```python
rmse_dtr = root_mean_squared_error(y_test, predicciones_dtr)
print(f"âœ… Decision Tree Model RMSE: {rmse_dtr:.2f}")
```

```python
margen_error_dtr = (rmse_dtr / y_test.mean()) * 100
print(margen_error_dtr)
```

```python
#-------------RandomForest-----------------
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
randomForest_model = RandomForestRegressor(n_estimators=100, random_state=42)
randomForest_model.fit(X_train, y_train)
```

```python
predicciones_rfr = randomForest_model.predict(X_test)
```

```python
rmse_rfr = root_mean_squared_error(y_test, predicciones_rfr)
print(f"âœ… Random Forest Model RMSE: {rmse_rfr:.2f}")
```

```python
margen_error_rfr = (rmse_rfr / y_test.mean()) * 100
print(margen_error_rfr)
```

```python
import pandas as pd

# Build performance comparison DataFrame for Decision Tree, Random Forest, and XGBoost
tree_models_df = pd.DataFrame({
    'Model': ['Decision Tree', 'Random Forest', 'XGBoost'],
    'RMSE': [rmse_dtr, rmse_rfr, rmse],
    'Relative Error (%)': [margen_error_dtr, margen_error_rfr, margen_error]
})

# Display the results
print("ðŸŒ² Tree-Based Regression Models Performance:")
display(tree_models_df)
```
